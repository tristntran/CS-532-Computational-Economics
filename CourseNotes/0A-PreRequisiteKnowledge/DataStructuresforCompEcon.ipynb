## List, Arrays, and Grids

Lists, arrays, and grids are all abstracted in python and implemented in the same way.
A list is what it sounds like and contains a list of values or variables or objects that can be iterated over, indexed, mapped onto etc. Here are some great examples.

### Creating a list

We can create lists in python by using square brackets. Anything between square brackets will be an item of that list. Here are some examples of creating lists

```
empty_list = []
old_list = [1,2,3]
new_list = [3,2,1]
```

### Advanced skill: List comprehension
We can create lists through for loops using list comprehensions. 

```
easy_list = [i for i in range(10)]
easy_list
```

```
list_comp = [i for i in range(10) if not i%5]
list_comp
```

### Comparing Lists
Just like otehr variables we can use the `==` operator to see if two lists have equivalent contents and order. 

```
new_list == old_list
```

```
new_list is old_list
```
