# Permutation: When order matters

Permutations are also called arrangements or orders. Let's start with a string "abcd". How many ways can we order these characters? Let's think of this problem as the following. We have four balls. One labeled A, one labeled B, one labeled C, and one labeled D.

We also have 4 bins. 1, 2, 3, and 4.

Our goal is to find out how many four letter words we can write. Using our example of balls and bins, we can imagine ourselves drawing balls out of a bag and randomly placing them in the bins. Our problem will have two different answers depending on whether or not we are allowed to replace the balls after they are drawn.

## Permutation with Replacement
Let's say that after we draw a ball from the bag, we write down the answer and then put it back in the bag. How many different "words" will we get?

To answer this problem let's answer an easier problem.

What if there were three tosses? Four? Five? What about k tosses?

For k tosses and 2 possible outcomes for each toss, the total number of possible outcomes will be $2^k$

## Permutation without replacement

When mathematicians think of permutation, they usually are talking about permutation without replacement. Let's take our example of "ABCD" we wan to know how many ways we can reorder the characters without any repeats.

Back to our example with bins and balls. Let's say that we are drawing the balls from a bag, and placing them into the bins without replacing them.

\___   \___   \___   \___

Now how many was are there to fill the first bin? There are four letters, so there are four possibilities.

\_4_   \___   \___   \___

How many ways are there to fill the second bin? You can fill it with any three of the remaining letters.

\_4_   \_3_   \___   \___

and so on and so fourth. Using our multiplication trick we can write this as $4\cdot 3\cdot 2\cdot 1 = 24$ permutations.

This is easy with small numbers under 10, but gets out of hand really fast. To multiply a sequence of natural numbers from 1 to n we can write the factorial function

>$n! = 1 \cdot 2 \cdot ... \cdot n$

>The number of ways to reorder or permute set of n items is n!.

What if we want to use the entire alphabet now. How many four letter "words" can we create without repeating any letters?

We still have four bins, but instead have 26 ways to draw the first letter. The second bin can be filled with anything except the used character (25 ways).

This means that there are $26 \cdot 25 \cdot 24 \cdot 23$ different words we can write.

Let's generalize this to a set of n items of length k without repitition where order matters. We have n ways to choose the first letter, n-1 ways to choose the second and so on and so forth.

$n \cdot (n-1) \cdot (n-1) \cdot ... \cdot (n-k+1)$

Let's rewrite this using factorials. notice that this is just $n!$ but we ignore all terms that are less than or equal to (n-k) so we can write this as:  $\frac{n!}{(n-k)!}$

> The number of length k<n ordered sets from a list of n items without repitition is $\frac{n!}{(n-k)!}$

We are going to first need to recognize the pattern of mutliplying consecutive integers as a factorial. It is usually denoted with the operator "!". 

$n! = n*(n-1)*(n-2)*...*3*2*1$

if you would like feel free to play around with the below code that will calculate the factorial and print out the result. Not the best code, but it shows what it needs to show.


```python
def factorial(n):
    #calculates factorial
    #verbose
    fact = 1
    s = ""
    for i in range(1,n+1):
        fact *= i
        s += " {} *".format(i)
    print("{}! = ".format(n)+s[:-1] + " = {}".format(fact))

factorial(3)
```

So what if we only want to multiply from 16 to 14? We are going to take 16! and divide 13! 

$16 \cdot 15\cdot14 = \frac{16!}{13!}$

Another way to look at it is by providing an alternative process by which we choose the three billiards. Imagine we have all 16 billiards lined up and then shuffle them into all of their possible permutations. There are 16! of these.

We will only pay attention to the first three billiards.

This means that we will have 16! possible arrangements, but since only the first 3 matter, the other 13 don't.

We need to divide out the repeats. For each arrangement of the first three, we can rearrange the remaining billiards 13! different ways. We are counting each arrangement of the first three 13! times. To avoid double, triple, or 13!-ple counting we will divide the total number of permutations by the permutations of the excess. Our final answer is $\frac{16!}{13!}$

# Combination: When order doesn't matter:

Let's say that we want to find out how many groups k we can form from n people. Or let's say we want to find out what combinations of toppings we can put on our pancakes. We don't care what order these toppings are placed, instead we care about inclusion or exclusion: how many **unique** sets there are.

We will take our equation from earlier $\frac{n!}{(n-k)!}$ and divide it by the ways to permute our list of k items

$\frac{n!}{(n-k)!}\div k! =\frac{n!}{(n-k)!k!}$

This problem is so common that it has its own notation. We read this as n choose k.

$_nC_k = {{n} \choose {k}} =\frac{n!}{(n-k)!k!}$

## 4. Combinations with reptition allowed (Stars and Bars):

This might not be intuitive at first, but what you're doing is choosing one out of n and then putting it back in the mix. You choose one and then put it back in. We will replace our selection every single time we choose something except for our last one. 

Try the following example.

Let's say there are 5 ice cream flavors how many scoop variations of 3 can you have?
Order does not matter, and we are allowed to repeat. Before reading onwards try to create your own solution.

Here is a bit of code that might help. How many combinations would it be if you had n flavors to choose from and k scoops?

```python
def iceCreamCombos(flavors):
    combos ={}
    for i in flavors:
        for j in flavors:
            for k in flavors:
                combo = tuple(sorted([i,j,k]))
                if combo in combos:
                    combos[combo] += 1
                else:
                    combos[combo]= 1
    print(len(combos))
    for combo in combos:
        print("{} \t : \t {}".format(combo,combos[combo]))
    return 0

flavs = {"Vanilla","Chocolate","Strawberry","Rocky Road","Cookie Dough"}
iceCreamCombos(flavs)
```



We can approach this problem by representing it as a tuple. We have five flavors so it will be a tuple of length 5.

The sum of the tuple must add up to 3.

(1,0,1,1,0) represents one of the first flavor, none of the second, one of the third, one of the fourth, and none of the fifth.

Representing distributions and strategies in the form of tuples is going to be a recurring theme in Computational Economics.

This type of problem is commonly referred to as stars and bars. When ever you are looking to find a non-negative tuple of length n that adds up to equal a certain number k, we will imagine this as having k stars and n bars.

We will use the bars to divide the stars into groups how many positions can we fit the bars into?

## \* \* \*

We can see that there are four gaps represented by the horizontal lines. 
##\_\*\_\*\_\*\_

We want to group these stars into 5 different bins corresponding to the tuple.

## \*\|\|\*\|\*\|

The above corresponds to our earlier tuple. (1,0,1,1,0)

So how many objects do we have total? The answer is we have k stars and n-1 bars. So we have a total of n+k-1 objects total. We can rearrange these in (n+k-1)! different ways (permutation); however, all of the stars are identical and all of the bars are identical when we read and interpret them. So we need to divide our permutation by the permutation of the stars and the bars. $\frac{(n+k-1)!}{(n-1)!(k)!}=\frac{(n+k-1)!}{(n)!(k-1)!} = {{n+k-1} \choose {k-1}} = {{n+k-1} \choose {n}}$

There's a lot of ways to write this one, if you're not convinced, expand the factorial and you will see how they are equivalent. You can also test it with some small numbers.

For our ice cream example with 5 flavors and 3 scoops there are ${{5+3-1} \choose {5}}= 21$ different combinations.
